```python
# File: main.py atau Cell Baru di Jupyter Notebook

# Defining the Handphone class directly in this cell to resolve ImportError
class Handphone:
    def __init__(self, merk, model, harga, stok):
        self.merk = merk
        self.model = model
        self.harga = harga
        self.stok = stok

    def tampilkan_info(self):
        print(f"Merk: {self.merk}, Model: {self.model}, Harga: Rp{self.harga:,}, Stok: {self.stok}")

    def kurangi_stok(self, jumlah):
        if self.stok >= jumlah:
            self.stok -= jumlah
            return True
        else:
            return False

def hitung_total(harga, jumlah):
    """Fungsi sederhana untuk menghitung total belanja (Konsep Modularisasi Fungsi)"""
    return harga * jumlah

def main():
    # 1. Membuat objek (Kelas Sederhana)
    hp_a = Handphone("Xiaomi", "Redmi Note 13", 2500000, 10)
    hp_b = Handphone("iPhone", "15 Pro", 18000000, 2)

    print("=== TOKO HP MAJU JAYA ===")
    hp_a.tampilkan_info()
    hp_b.tampilkan_info()

    # 2. Simulasi Penjualan
    pesanan = 3
    print(f"\n--- Transaksi: Pembelian {pesanan} unit {hp_a.model} ---")

    if hp_a.kurangi_stok(pesanan):
        total = hitung_total(hp_a.harga, pesanan)
        print(f"Transaksi Berhasil!")
        print(f"Total Bayar: Rp{total:,}")
    else:
        print("Transaksi Gagal: Stok tidak cukup!")

    print("\n=== UPDATE STOK TERBARU ===")
    hp_a.tampilkan_info()

if __name__ == "__main__":
    main()
```
